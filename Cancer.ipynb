```python
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split

dataset = pd.read_csv("cancer.csv")
```

```python
data = dataset.drop(columns=["diagnosis(1=m, 0=b)"])
diag = dataset["diagnosis(1=m, 0=b)"]
```

```python
trainData, testData, trainDiag, testDiag = train_test_split(data, diag, test_size=0.2)
md = tf.keras.models.Sequential()
```

```python
md.add(tf.keras.layers.Dense(256, input_shape=trainData.shape, activation='sigmoid'))
md.add(tf.keras.layers.Dense(256, activation='sigmoid'))
md.add(tf.keras.layers.Dense(1, activation='sigmoid'))
```

```python
md.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
md.fit(trainData, trainDiag, epochs=1000)
```

```python
md.evaluate(testData, testDiag)
```
